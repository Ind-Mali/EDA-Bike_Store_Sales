# Observations

* There are mainly adults in the dataset and the most important incomes belong to adults.
* Customer gender is close to each other.
* Australia and USA are the countries with the highest sales.
* The top selling states are California, British Columbia, England, Washington and New South Wales.
* Most items were sold in accessory products(1054162 Quantity) , but most revenue was derived from bicycle sales(61782134 Revenue).
* Although the product with the most sales was Tires and Tubes(514051 Order Quantity), the most income was obtained from Road Bikes(33363061 Revenue).
* Although the age of 31 is the highest order in terms of units, the highest income is made by 28-year-olds.
* Although the most orders on the basis of units were in 2014 and 2016, the highest revenue was made in 2015 and 2016.
* On the basis of units, the most orders were made in December and the least orders were made in July.
* The most income was obtained on the 24th of the month.

# Label Encoder
We have lots of unique items, but some of them are categorical. We should change these categorical values because the computer does not understand.

```python
from sklearn.preprocessing import LabelEncoder
dfLabel = dfSalas.apply(lambda x : LabelEncoder().fit_transform(x) if x.dtype =="O" else x)
dfLabel
```

# Correlation Test

```python
mask = np.zeros_like(dfLabel.corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices]=True
```

```python
colNames = list(dfLabel[1:])
colNames
```

```python
comb = list(combinations(colNames,2))
binaSelect = [list(i) for i in set(tuple(i) for i in comb)]
binaSelect
```

```python
for i in range(0, len(binaSelect)):
    dfSalasTemp = dfLabel[binaSelect[i]]
    print(dfSalasTemp.columns[0] + "|" + dfSalasTemp.columns[1], "====>>>>> P_Value : ", 
          scipy.stats.pearsonr(dfSalasTemp.iloc[:,0], dfSalasTemp.iloc[:,1])[1])
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(dfLabel.corr("pearson"), annot= True, mask=mask)
```
